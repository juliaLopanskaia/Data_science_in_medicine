```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(x, y)

Z1 =  X - 0.1*Y
Z2 =  X - 0.1*Y + 30

ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

x = np.arange(-50, 50, 0.5)
y = np.arange(-50, 50, 0.5)
X, Y = np.meshgrid(x, y)

Z2 =  np.sqrt(100000 - X**2 - Y**2)


ax.plot_wireframe(X, Y, Z2)
show()
```
