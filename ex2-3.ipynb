```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10, 1000)
R = 10

y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)


plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.linspace(-10,10, 1000)
a = 10
b = 20

y1 = np.sqrt((1 - (x/a)**2)*b**2)
y2 = -np.sqrt((1 - (x/a)**2)*b**2)


plt.plot(x, y1)
plt.plot(x, y2)
```

```python
a = 10
b = 20
x1 = np.linspace(a,100, 100)

y1 = np.sqrt((-1 + (x1/a)**2)*b**2)
y2 = -np.sqrt((-1 + (x1/a)**2)*b**2)

y = []
y.extend(list(-y1))
y.extend(list(-y2))
y.extend(list(y1))
y.extend(list(y2))
x = []
x.extend(list(-x1))
x.extend(list(-x1))
x.extend(list(x1))
x.extend(list(x1))



plt.scatter(x, y)
```
