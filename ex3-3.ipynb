```python
import numpy as np

alpha = np.pi/2
r = 1

x = r * np.cos(alpha)
y = r * np.sin(alpha)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

r = 1
alpha = np.arange(-np.pi, 3/2*np.pi, np.pi/50)


plt.plot(r * np.cos(alpha), r * np.sin(alpha))
```

```python
import matplotlib.pyplot as plt
import numpy as np

r = 1
alpha = np.arange(-np.pi, 3/2*np.pi, np.pi/50)


plt.plot(r * np.cos(alpha),2*(r * np.cos(alpha)))
```
